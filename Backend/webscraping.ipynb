```python
from urllib.request import urlopen
import re
import os
import requests
from urllib.parse import unquote
```

```python
url = "https://rca.alaska.gov/RCAWeb/RCALibrary/RCAAnnualReportList.aspx"
```

```python
page = urlopen(url)
```

```python
page.geturl()
```

```python
html_bytes = page.read()
html = html_bytes.decode("utf-8")
```

```python
html
```

```python
# Gets all <a href> HTML

# pattern = "<a href.*?</a.*?>"
# a_tags = re.findall(pattern, html)
# a_tags
```

```python
# Gets only pdf links
pattern = "<a href.*?pdf</a.*?>"
a_tags = re.findall(pattern, html)
a_tags
```

```python
# Create the PDFs directory if it doesn't exist
if not os.path.exists('PDFs'):
    os.makedirs('PDFs')

# Download each PDF
for link in a_tags:
    # Extract URL and filename
    url = link.split('href="')[1].split('"')[0]
    filename = link.split('>')[1].split('<')[0]  # Get the text between > and < as the filename

    # Download the file
    response = requests.get(url)
    path = os.path.join('PDFs', filename)  # Create a valid path for the file
    with open(path, 'wb') as f:
        f.write(response.content)
    print(f'Downloaded {filename} into PDFs/')

print('DONE downloading PDFs')
```
